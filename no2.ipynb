最近主流のJitterプロットを重ね合わせるには、numpy.random の normal() 関数でランダムな値を生成し、その分だけ左右にずらして plot() 関数で点を重ね合わせます

```
plt.style.use('ggplot')
plt.figure()
x1, x2, x3 = iris[0:50, 0], iris[50:100, 0], iris[100:150, 0]
data = [x1, x2, x3]
plt.boxplot(data, labels=['Setosa', 'Versicolor', 'Virginica'])
for i in range(len(data)):
  y = data[i]
  x = np.random.normal(1+i, 0.04,size = len(y))
  plt.plot(x, y,'.k', alpha=.8)
plt.show()
```

#相関

分散共分散行列を求めるには、numpy の cov() 関数を使います。



```
x, y = iris[0:50, 0], iris[0:50, 1]
cov = np.cov(x, y)
print('分散共分散行列:', cov)
print('共分散:', cov[0, 0])
```

偏差の符号が同じ（平均を表す点より右上と左下の領域にある）値が多ければ共分散は大きくなり、偏差の符号が異なる（平均を表す点より左上と右下の領域にある）値が多ければ共分散は小さくなります。

#相関係数、相関係数行列

相関係数行列(２つの系列の似ている度合いを測る尺度)を求めるには、numpy の corrcoef() 関数を使います。

```
x, y = iris[0:50, 0], iris[0:50, 1]
r = np.corrcoef(x, y)
print('相関係数行列:', r)
print('相関係数:', r[0, 1])
```

corrcoef() 関数に行列（2次元のndarray）を与えることにより、行が変数、列が観測を表すデータとして相関係数行列を求めることができます。

irisの元データは、行が観測、列が変数を表していますので、rowvar=False オプションを追加するか、行列を転置する必要があります。

```
np.corrcoef(iris[0:50], rowvar=False)
```

```
np.corrcoef(iris[0:50].T)
```

相関係数の値によって、相関を次のように表現します。
（ただし、この値の範囲や表現は厳格に決まっているわけではありません。）
*   $R \ge 0.9$のとき「極めて強い正の相関がある」
*   $0.7 \le R \le 0.9$のとき「強い正の相関がある」
*   $0.4 \le R \le 0.7$のとき「（中程度の）正の相関がある」
*   $0.2 \le R \le 0.4$のとき「弱い正の相関がある」
*   $-0.2 \le R \le 0.2$のとき「ほとんど相関がない」
*   $-0.4 \le R \le -0.2$のとき「弱い負の相関がある」
*   $-0.7 \le R \le -0.4$のとき「（中程度の）負の相関がある」
*   $-0.9 \le R \le -0.7$のとき「強い負の相関がある」
*   $R < -0.9$のとき「極めて強い負の相関がある」

定義より、$R_{i,j} = R_{j,i}$ が成り立ちます。

$R_{i,j} = R_{j, i} > 0$ で正の相関があるとき、「$X_i$が大きい**と**$X_j$も大きい」という関係と「$X_j$が大きい**と**$X_i$も大きい」という関係がどちらもあり、「$X_i$が大きい**から**$X_j$も大きい」または「$X_j$が大きい**から**$X_i$も大きい」という**因果関係**を表しているわけではないことに注意しましょう。

また、標本が小さいと、偶然に強い相関がある（または、ほとんど相関がない）ことがあります。

# 演習
今回の演習ではボストンの住宅価格データを使います。

このデータセットの `data` と `target` を連結して、データを作成します。
このデータには、以下の値が含まれています。
0. CRIM per capita crime rate by town
1. ZN proportion of residential land zoned for lots over 25,000 sq.ft.
2. INDUS proportion of non-retail business acres per town
3. CHAS Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
4. NOX nitric oxides concentration (parts per 10 million)
5. RM average number of rooms per dwelling
6. AGE proportion of owner-occupied units built prior to 1940
7. DIS weighted distances to five Boston employment centres
8. RAD index of accessibility to radial highways
9. TAX full-value property-tax rate per \$10,000
10. PTRATIO pupil-teacher ratio by town
11. B 1000(Bk - 0.63)^2 where B_k is the proportion of blacks by town
12. LSTAT % lower status of the population
13. MEDV Median value of owner-occupied homes in \$1000’s

```
import numpy as np
from sklearn.datasets import load_boston

boston_dataset = load_boston()
length = len(boston_dataset.target)
boston = np.concatenate((boston_dataset.data, boston_dataset.target.reshape(length,1)), axis=1)
print(boston)
```

```
x = boston[:,13]
print("平均:", np.mean(x))
print("中央値:", np.median(x))
print("最頻値:", mode(x)[0][0])
```

```
a = np. array(x)
v = np.var(x)
sigma = np.std(x)
print("標準偏差:", sigma)
```
